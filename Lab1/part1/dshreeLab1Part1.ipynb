# Problem 1

Plot line graph. Specification used:
1) par:Set Graphical parameters of Plot
2) type: type of plot(o: overplot)
3) ylim: range limit of yaxis
4) xlab, ylab: Labek of X and Y axis
5) inset: Fraction Distance from margin legend to put
6) pch: plot Character
7) lty: line Type(2: dash)

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)#Create list of numbers
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk") #set background color

#Plot graph with sales1
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")#Title of the graph

#Plot New Lie for sale2
lines(sales2, type="o", pch=22, lty=2, col="red")

#Add grid with number of cells corresponding to tick marks on x and y axis
grid(nx=NA, ny=NULL)

#Add legend to the graph
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

# Problem 2

Plot a bargraph from text file. Specifications used:
1) main: Title to barplot
2) beside: draw bars for sales side by side
3) col: use rainboe color palette

```r
#Read data from txt file
sales<-read.table("salesdata.txt", header=T)
sales  # to verify that data has been read
#Convert input to matrix and plot bargraph
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(5))
```

# Problem 3

Create a boxplot for the Sales data read above

```r
fn<-boxplot(sales,col=c("orange","green"))$stats

#Add text as Median of Sales1 and Sales 2
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

# Problem 4

Draw Histogram and Line Chart for Apple and Facebook Stock

```r
#Read CSVs to dataframe
fb1<-read.csv("FB.csv")
aapl1<-read.csv("AAPL.csv")
#set background color
par(bg="cornsilk")
plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(150,200), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)

#Draw histogram with rainbow colors
hist(aapl1$Adj.Close, col=rainbow(8))
```
